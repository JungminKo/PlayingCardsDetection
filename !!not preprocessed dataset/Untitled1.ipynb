```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
image = cv2.imread('PlayingCards_52.png')
# image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
im_bw = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
thresh, im_bw = cv2.threshold(im_bw, 170, 255, 0)
contours, hierarchy = cv2.findContours(im_bw.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
rects = [cv2.minAreaRect(each) for each in contours]
rects = [(a,b,c) for a, b, c  in rects if ((b[0]*b[1]>5000))]
```

```python
tmp=1
for i in rects:
    box = cv2.boxPoints(i)
    box = box.astype('int')
    cv2.imwrite('deck/'+str(tmp)+'.jpg', image[box[1][1]:box[0][1],box[0][0]:box[2][0], :])
    tmp+=1
```

```python
for i in os.listdir('deck'):
    image = cv2.imread('deck/'+i)
    image = cv2.cv2.resize(image, dsize=(100, 120), interpolation=cv2.INTER_AREA)
    cv2.imwrite('deck/'+i, image)
```

```python
image = cv2.imread('PlayingCards_s.jpg')
im_bw = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
# img_sobel_x = cv2.Sobel(im_bw, cv2.CV_64F, 0, 1, ksize=1)
img_sobel_x = cv2.convertScaleAbs(im_bw)
im_bw = cv2.Canny(img_sobel_x, 170, 200)
thresh, im_bw = cv2.threshold(im_bw, 170, 255, 0)
contours, hierarchy = cv2.findContours(im_bw.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

rects = [cv2.minAreaRect(each) for each in contours]
rects = [(a,b,c) for a, b, c  in rects if ((b[0]*b[1]>35000) & (b[0]*b[1]<1000000))]
img = image.copy()
col = 0
for rect in rects:
#     col +=20
    box = cv2.boxPoints(rect)
    box = box.astype('int')
    img = cv2.drawContours(img, [box], -1, (255, 0,0), 3)
plt.figure(figsize=(20,10))
plt.imshow(img)
```
