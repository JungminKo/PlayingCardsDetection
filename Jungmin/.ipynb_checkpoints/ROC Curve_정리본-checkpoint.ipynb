```python
import math
import numpy as np
import cv2, os
import pandas as pd
from numpy import linalg as LA
from PIL import Image
import matplotlib.pyplot as plt
```

```python
def face_matrix():
    S = []
    Y = []

    for name in os.listdir("./ROC/train"):
        nameList = os.listdir("./ROC/train/%s"%name)

        if len(nameList) <50 : 
            continue

        y = name.replace(".jpg","")

        count = 0
        for i in nameList:
            if count == 50 :
                break
            filePath = "./ROC/train/%s/%s"%(name,i)

            # 한글 경로 디코딩
            stream = open( filePath.encode("utf-8") , "rb")
            bytes = bytearray(stream.read())
            numpyArray = np.asarray(bytes, dtype=np.uint8)
            img = cv2.imdecode(numpyArray , cv2.IMREAD_UNCHANGED)

            gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
            imgArr = np.array(gray)
            temp = np.reshape(imgArr, (150*150))
            Y.append(y)
            S.append(temp)
            count += 1

    S = np.array(S).T
    return S , Y

def histo_equalization(S):
    num0 = S.shape[0]
    num1 = S.shape[1]
    SS = np.zeros((num0, num1))
    
    for i in range(num1):
        equ = cv2.equalizeHist(S[:,i]).reshape(22500,)
        SS[:,i] = equ
    
    m = np.mean(SS, axis=1)
    m = m.reshape(-1, 1)
    D = SS-m
    return D, m

def eigenface(D, eigenface_num=20):
    L = np.dot(D.T, D)
    w, v = LA.eig(L)


    # 고유값 0인거 제거    
    zero_eigen = [i for i in range(len(w)) if abs(w[i]) <1e-4]
    j=0
    for i in zero_eigen:
        w = np.delete(w, i-j, axis=0)
        v = np.delete(v, i-j, axis=1)
        j+=1

    # 고유값 내림차순으로 정렬
    ww = np.sort(w.real)[::-1]
    index = np.argsort(w.real)[::-1]  
    vv = np.array([v[:,i].real for i in list(index)]).T 

    ww = ww[:eigenface_num]
    vv = vv[:, :eigenface_num]

    u = np.dot(D, vv)
    u = u/(LA.norm(u, axis=0))

    weight=np.dot(u.T,D)
    return u, weight

def fisherface(D, y, eigenface_num=30, fisherface_num=15):
    u, weight = eigenface(D, eigenface_num)
    
    weight_T = weight.T
    all_mean = weight_T.mean(axis=0)
    SID = np.unique(y)
    y = np.array(y)
    
    n = weight_T.shape[1]
    Sw = np.zeros((n, n), dtype=np.float32)
    Sb = np.zeros((n, n), dtype=np.float32)
    
    for i in SID:
        SID_bet_weight = weight_T[np.where(y==i)]
        SID_mean = SID_bet_weight.mean(axis=0)
        num_per_SID = SID_bet_weight.shape[0]
        
        for j in range(num_per_SID):
            diff_w = (SID_bet_weight[j]-SID_mean).reshape(-1,1)
            Sw = Sw+np.dot(diff_w, diff_w.T)
        diff_b = (SID_mean-all_mean).reshape(-1, 1)
        Sb = Sb+ (num_per_SID)*np.dot(diff_b, diff_b.T)
    
    m = np.linalg.inv(Sw)*Sb
    w, v = LA.eig(m)
    
    zero_eigen = [i for i in range(len(w)) if abs(w[i]) <1e-4]
    j=0
    for i in zero_eigen:
        w = np.delete(w, i-j, axis=0)
        v = np.delete(v, i-j, axis=1)
        j+=1
    
    # 고유값 내림차순으로 정렬
    ww = np.sort(w.real)[::-1]
    index = np.argsort(w.real)[::-1]  ### 한번에 가져오는 방법 없는지 확인
    vv = np.array([v.real[:,i] for i in list(index)]).T 
    
    ww = ww[:fisherface_num]
    vv = vv[:, :fisherface_num]
    
    u_fisher = np.dot(u,vv)
    weight_fisher=np.dot(u_fisher.T,D)
    
    return  u_fisher, weight_fisher

def cosine_recognition(img, mean_face, eigenvector, SID_weight, SID_index):
    img = cv2.equalizeHist(img) - mean_face
    img_weight = np.dot(img.T,eigenvector)
    dist = np.sqrt(np.sum(img_weight*img_weight))
    
    # cosine_similiarity= pd.DataFrame(columns = y)
    cosine_similiarity = np.zeros((1, SID_weight.shape[0]))
    for i in range(SID_weight.shape[0]):
        cosine_similiarity[0, i]=(np.sum(img_weight*SID_weight[i]))/((dist)*(np.sqrt(np.sum(SID_weight[i]*SID_weight[i]))))
    
    cosine_similiarity = pd.DataFrame(cosine_similiarity, columns=SID_index)
    return cosine_similiarity

def ROC_dir(eigen = True, num1=30, num2=15):
    S, y = face_matrix()
    D, m = histo_equalization(S)
    if eigen==True:
        vector, weight = eigenface(D,num1)
    
    else:
        vector, weight = fisherface(D, y, num1, num2)
    
    labels = os.listdir("./ROC/test")
    cols = y.copy()
    cols.append("label")
    cosine_similiarity= pd.DataFrame(columns = cols)

    for label in labels:
        for imgs in os.listdir("./ROC/test/%s/"%label):
            img = cv2.imread("./ROC/test/%s/%s"%(label,imgs))
            gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
            temp = np.reshape(gray, (150*150)) 

            cosine_similiarity1 = cosine_recognition(temp, m, vector, weight.T, y)
            cosine_similiarity1["label"]=label
            cosine_similiarity= pd.concat([cosine_similiarity, cosine_similiarity1])

        print("./ROC/test/%s/%s"%(label,imgs))
    #cosine_similiarity.to_csv("./ROC/cosine_similiarity.csv", mode='w')

    return cosine_similiarity

def plot(eigen=True, num1=20, num2=10):
    cosine_similiarity = ROC_dir(eigen, num1)
    person = np.unique(cosine_similiarity["label"])
    num_person = len(person)
    authentic = np.zeros((num_person*50*50,))
    imposter = np.zeros((num_person*50*50*(num_person-1),))
    fig = plt.figure(figsize=(7,7))
    a = np.ceil(num_person/4)
    j=0
    for i in range(num_person):
        name = person[i]

        for naname in person:
            T_T = np.array(cosine_similiarity[cosine_similiarity["label"]==name].loc[:, naname:naname]).reshape(-1,)
            if naname == name:
                temp=i*2500
                authentic[temp:temp+2500] = T_T
            else:
                temp = j*2500
                imposter[temp:temp+2500] = T_T
                j+=1 

        
        ax = fig.add_subplot(4,a, i+1)
        ax.set_title(name)
        plt.hist(authentic[i*2500:(i+1)*2500], range=[-1,1], color = 'r',  bins=100, alpha = 0.6)
        plt.hist(imposter[i*2500*6:(i+1)*2500*6], range=[-1,1], color = 'b',  bins=100, alpha = 0.6 )
        plt.tight_layout()

    fig = plt.figure(figsize=(7,5))
    plt.ylim((0, 4000))
    plt.hist(authentic, range=[-1,1],   color = 'r',  bins=100, alpha = 0.6, label = "authentic")
    plt.hist(imposter, range=[-1,1],   color = 'b',  bins=100, alpha = 0.6, label = "imposter")
    plt.axvline(x=authentic.mean(), color = 'r')
    plt.axvline(x=imposter.mean(), color = 'b')
    plt.legend()
    if eigen==True:
        title = './ROC/eigen_'+str(num1)+'_full_191125.png'
    else:
        title = './ROC/fisher_'+str(num1)+'_'+str(num2)+'_full_191125.png'
    # plt.savefig(title, dpi=300)
    
    return  authentic, imposter # authentic.mean(), authentic.std(), imposter.mean(), imposter.std()

def FAR_FRR (authentic, imposter):#, name='./ROC/ROC.png'):
    threshold_array = np.arange(-1,1,0.01)
    FAR_FRR_len = np.zeros((len(threshold_array),3))
    
    i=0
    auth_len = len(authentic)
    impo_len = len(imposter)
    for threshold in threshold_array:
        FAR_FRR_len[i][0]=threshold
        FAR_FRR_len[i][1]=len(authentic[authentic<threshold])/auth_len
        FAR_FRR_len[i][2]=len(imposter[imposter>threshold])/impo_len
        i+=1
    index = np.argmin(np.abs(FAR_FRR_len[:,1]-FAR_FRR_len[:,2]))
    threshold_EER = FAR_FRR_len[index][0]
    EER = np.mean(FAR_FRR_len[index][1:3])
    
    return threshold_EER,  EER, FAR_FRR_len
```

```python
authentic_5, imposter_5 = plot(eigen=True, num1=5)
threshold_EER_5,  EER_5, FAR_FRR_len_5 = FAR_FRR(authentic_5, imposter_5)
```

```python
authentic_10, imposter_10 = plot(eigen=True, num1=10)
threshold_EER_10,EER_10, FAR_FRR_len_10 = FAR_FRR(authentic_10, imposter_10)
```

```python
authentic_20_5, imposter_20_5 = plot(eigen=False, num1=20, num2=5)
threshold_EER_20_5,EER_20_5, FAR_FRR_len_20_5 = FAR_FRR(authentic_20_5, imposter_20_5)
```

```python
authentic_50_10, imposter_50_10 = plot(eigen=False, num1=50, num2=10)
threshold_EER_50_10,EER_50_10, FAR_FRR_len_50_10 = FAR_FRR(authentic_50_10, imposter_50_10)
```
