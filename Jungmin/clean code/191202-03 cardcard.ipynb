```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def edge_detect(file_name, tresh_min, tresh_max):
    f = plt.figure(figsize=(20,10))
    ax = f.add_subplot(131)
    ax2 = f.add_subplot(132)
    ax3 = f.add_subplot(133)
    
    image = cv2.imread(file_name)
    im_bw = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
    thresh, im_bw = cv2.threshold(im_bw, tresh_min, tresh_max, 0)

    
#     cv2.imwrite('bw_'+file_name, im_bw)

    contours, hierarchy = cv2.findContours(im_bw, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
    ax.imshow(image, cmap='Greys')
    cv2.drawContours(image, contours, -1, (0,255,0), 3)
    
    ax2.imshow(im_bw)
    ax3.imshow(image)
    return contours
#     cv2.imwrite('cnt_'+file_name, image)
```

### 라플라시안 필터

```python
image = cv2.imread('heart3.jpg')
im_bw = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
laplacian = cv2.Laplacian(im_bw,cv2.CV_64F)
plt.figure(figsize=(10,5))
plt.imshow(laplacian, cmap='gray')
```

### Sobel 필터

```python
img = cv2.imread('spade4.jpg',cv2.IMREAD_GRAYSCALE)
canny = cv2.Canny(img,30,70)

laplacian = cv2.Laplacian(img,cv2.CV_8U)
sobelx = cv2.Sobel(img,cv2.CV_64F,1,0,ksize=15)
sobely = cv2.Sobel(img,cv2.CV_64F,0,1,ksize=15)

images = [img,laplacian, sobelx, sobely, canny]
titles = ['Origianl', 'Laplacian', 'Sobel X', 'Sobel Y','Canny']

for i in range(5):
    plt.figure(figsize=(20, 10))
    plt.plot(),plt.imshow(images[i]),plt.title([titles[i]])
    plt.xticks([]),plt.yticks([])

plt.show()
```

### Contour 사용

```python
image = cv2.imread('heart3.jpg')
im_bw = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
thresh, im_bw = cv2.threshold(im_bw, 170, 255, 0)

contours, hierarchy = cv2.findContours(im_bw.copy(), cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

rects = [cv2.boundingRect(each) for each in contours]
rects = [(x,y,w,h) for (x,y,w,h) in rects if ((w*h>1000))]

for rect in rects:
    # Draw the rectangles
    cv2.rectangle(image, (rect[0], rect[1]), 
                  (rect[0] + rect[2], rect[1] + rect[3]), (255, 0, 0), 5) 

plt.figure(figsize=(15,12))
plt.imshow(image)
```

```python
image = cv2.imread('spade4.jpg')
im_bw = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
thresh, im_bw = cv2.threshold(im_bw, 170, 255, 0)
contours, hierarchy = cv2.findContours(im_bw.copy(), cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

rects = [cv2.minAreaRect(each) for each in contours]
rects = [(a,b,c) for a, b, c  in rects if ((b[0]*b[1]>50000))]
img = image.copy()
for rect in rects:
    box = cv2.boxPoints(rect)
    box = box.astype('int')
    img = cv2.drawContours(img, [box], -1, (255, 0, 0), 5)
plt.figure(figsize=(20,10))
plt.imshow(img)
```

```python
image = cv2.imread('heart3.jpg')
im_bw = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
thresh, im_bw = cv2.threshold(im_bw, 170, 255, 0)
contours, hierarchy = cv2.findContours(im_bw.copy(), cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

rects = [cv2.minAreaRect(each) for each in contours]
a = np.array(rects)[:,1]
n = np.argmax([a[i][0]*a[i][1] for i in range(len(a))])
rect = rects[n]
img = image.copy()
box = cv2.boxPoints(rect)
box = box.astype('int')
img = cv2.drawContours(img, [box], -1, (255, 0, 0), 5)
plt.figure(figsize=(20,10))
plt.imshow(img)
```
