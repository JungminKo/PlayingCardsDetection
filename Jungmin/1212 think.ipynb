```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
img = cv2.imread('../au_dataset/C2_0_2509.jpg', cv2.IMREAD_GRAYSCALE)
plt.imshow(img, cmap = 'gray')
plt.show()
```

```python
_,img2 = cv2.threshold(img, 200, 255, cv2.THRESH_BINARY_INV)
plt.imshow(img2)
plt.show()
```

```python
plt.imshow(img2[:,:15])
```

```python
height, width = img2.shape
matrix = cv2.getRotationMatrix2D((width/2, height/2), 7, 1)
dst = cv2.warpAffine(img2, matrix, (width, height))
plt.imshow(dst)
```

```python
img = cv2.imread('./test.PNG', cv2.IMREAD_GRAYSCALE)
plt.imshow(img, cmap = 'gray')
plt.show()
```

```python
thresh, im_bw = cv2.threshold(img, 170, 255, 0)
contours, hierarchy = cv2.findContours(im_bw.copy(), cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

rects = [cv2.minAreaRect(each) for each in contours]
rects = [(a,b,c) for a, b, c  in rects if ((b[0]*b[1]>50000))]
img2 = img.copy()
for rect in rects:
    box = cv2.boxPoints(rect)
    box = box.astype('int')
    img2 = cv2.drawContours(img2, [box], -1, (0, 0, 0), 5)
plt.figure(figsize=(20,10))
plt.imshow(img2, cmap = 'gray')
```
