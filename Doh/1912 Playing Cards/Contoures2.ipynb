```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def edge_detect(file_name, tresh_min, tresh_max):
    f = plt.figure(figsize=(20,20))
    ax = f.add_subplot(311)
    ax2 = f.add_subplot(312)
    ax3 = f.add_subplot(313)
    
    image = cv2.imread(file_name)
    im_bw = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
    thresh, im_bw = cv2.threshold(im_bw, tresh_min, tresh_max, 0)

    
#     cv2.imwrite('bw_'+file_name, im_bw)

#     contours, hierarchy = cv2.findContours(im_bw, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    contours, hierarchy = cv2.findContours(im_bw, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
    ax.imshow(image, cmap='Greys')
    col = 0
#     print(len(contours))
#     for i in contours:
#         print(len(i))
#     cv2.drawContours(image, mylist, -1, (0,255,0), 3)
    cv2.drawContours(image, contours, -1, (0,255,0), 3)
#     for contour in contours:
#         cv2.drawContours(image, contour, -1, (0,col,0), 3)
#         col+20
    ax2.imshow(im_bw)
    ax3.imshow(image)
    cv2.imwrite('cnt_'+file_name, image)
    return contours
```
