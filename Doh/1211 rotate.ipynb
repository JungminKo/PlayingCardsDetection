```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
img = cv2.imread('../au_dataset/C2_0_2509.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(img, cmap = 'gray')
plt.show()
```

```python
_,img2 = cv2.threshold(img, 200, 255, cv2.THRESH_BINARY_INV)
plt.imshow(img2)
plt.show()
```

```python
plt.imshow(img2[:,:15])
```

```python
height, width = img2.shape
matrix = cv2.getRotationMatrix2D((width/2, height/2), 7, 1)
dst = cv2.warpAffine(img2, matrix, (width, height))
plt.imshow(dst)
```

```python
height, width = img2.shape
matrix = cv2.getRotationMatrix2D(((width-5), (height)-5), 6, 1)
dst = cv2.warpAffine(img2, matrix, (width, height))
plt.imshow(dst)
```

```python
M = np.float32([[1, 0, 5], [0, 1, -5]]) # 이미지를 왼쪽으로 5, 밑으로 1 이동시킵니다.
img_translation = cv2.warpAffine(img2, M, (width,height))
plt.imshow(img_translation)
plt.show()
```

```python
height, width = img_translation.shape
matrix = cv2.getRotationMatrix2D(((width), (height)), 6, 1)
dst = cv2.warpAffine(img_translation, matrix, (width, height))
plt.imshow(dst)
```

```python
img = cv2.imread('../au_dataset/C2_0_2509.jpg', cv2.IMREAD_GRAYSCALE)
M = np.float32([[1, 0, 1], [0, 1, -5]]) # 이미지를 왼쪽으로 5, 위으로 5 이동시킵니다.
img_translation = cv2.warpAffine(img, M, (width,height))
plt.imshow(img_translation,cmap = 'gray')
plt.show()
```

```python
height, width = img_translation.shape
matrix = cv2.getRotationMatrix2D(((width), (height)), 6, 1)
dst = cv2.warpAffine(img_translation, matrix, (width, height))
plt.imshow(dst,cmap = 'gray')
plt.show()
```
