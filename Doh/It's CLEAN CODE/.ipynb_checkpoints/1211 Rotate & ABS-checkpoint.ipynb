```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
img = cv2.imread('../../au_dataset/C2_0_2509.jpg', cv2.IMREAD_GRAYSCALE)
plt.imshow(img, cmap = 'gray')
plt.show()
```

```python
img.shape
```

```python
height, width = img.shape
M = np.float32([[1, 0, 1], [0, 1, -5]]) # 이미지를 왼쪽으로 1, 위으로 5 이동시킵니다. 근거는 5도만큼 돌렸을때 기준점을 옮겨준 것
img_translation = cv2.warpAffine(img, M, (width,height))
plt.imshow(img_translation,cmap = 'gray')
plt.show()
```

```python
height, width = img_translation.shape
matrix = cv2.getRotationMatrix2D(((width), (height)), 5, 1)
dst = cv2.warpAffine(img_translation, matrix, (width, height))
plt.imshow(dst,cmap = 'gray')
plt.show()
```

```python
C2 = cv2.imread('../../dataset/card_deck1/C2.jpg', cv2.IMREAD_GRAYSCALE)
cv2.absdiff(dst,C2).sum()
```

```python
height, width = img.shape
M = np.float32([[1, 0, 1], [0, 1, -5]]) # 이미지를 왼쪽으로 1, 위으로 5 이동시킵니다. 근거는 5도만큼 돌렸을때 기준점을 옮겨준 것
img_translation = cv2.warpAffine(img, M, (width,height))
plt.imshow(img_translation,cmap = 'gray')
plt.show()
```
