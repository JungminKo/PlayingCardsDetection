```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def edge_detect(image, tresh_min, tresh_max):
    f = plt.figure(figsize=(15,30))
    ax = f.add_subplot(131)
    ax2 = f.add_subplot(132)
    ax3 = f.add_subplot(133)
    
    im_bw = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
    thresh, im_bw = cv2.threshold(im_bw, tresh_min, tresh_max, 0)

    contours, hierarchy = cv2.findContours(im_bw, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
    ax.imshow(image, cmap='Greys')
    cv2.drawContours(image, contours, -1, (0,255,0), 3)
    
    ax2.imshow(im_bw, cmap='gray')
    ax3.imshow(image)
    return len(contours)
```

```python
img = cv2.imread('../cards.png')
# 사진, TRESH_MIN, TRESH_MAX
# 리턴값은 contours의 length
edge_detect(img, 110, 255)
```
